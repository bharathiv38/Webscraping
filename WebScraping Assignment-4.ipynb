Q2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#Importing requests
import requests

# importing regex
import re

from bs4 import BeautifulSoup
```

```python
def match_fixture(url): # defining function
    #initalizing driver & loading url
    driver = webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    # Getting internation match fixtures page
    driver.get(driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").get_attribute('href'))
    # Preparing soup & scraping data
    soup = BeautifulSoup(driver.page_source, 'lxml')
    match_title = [x.text for x in soup.find_all('span',class_ ='u-unskewed-text fixture__format')]
    series = [x.text for x in soup.find_all('span',class_ ='u-unskewed-text fixture__tournament-label u-truncated')]
    place = [x.text for x in soup.find_all('p',class_ ='fixture__additional-info')]
    date = [x.text for x in soup.find_all('span',class_ ='fixture__datetime tablet-only')]
    driver.close()
    # Preparing dataframe & cleaning data
    df= pd.DataFrame({})
    df['match_title']=match_title
    df['series']=series
    df['place1']=place
    df['date1']=date
    dfp = df.place1.str.split('\n',n = 4, expand = True)
    dfp.columns = ['d1','d2','place','df4']
    dfp.drop(['d1','d2','df4'],axis =1,inplace =True)
    dfd= df.date1.str.split('\n',n = 4, expand = True)
    dfd.columns = ['d1','date','time','d3','d4']
    dfd.drop(['d1','d3','d4'],axis =1,inplace =True)
    df = pd.concat([df,dfp,dfd],axis =1)
    df.drop(['place1','date1'], axis =1, inplace = True)
    return(df)


match_fixture('https://www.bcci.tv/') # Calling function
```

9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
def series(url):
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    name = [x.text for x in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")]
    year_span= [x.text for x in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")]
    genre = [x.text for x in driver.find_elements_by_xpath("//span[@class='genre']")]
    run_time = [x.text for x in driver.find_elements_by_xpath("//span[@class='runtime']")]
    ratings = [x.text for x in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")]
    votes = [x.text for x in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[@name='nv']")]
    driver.close()
    df = pd.DataFrame({})
    df['name'] =name 
    df['year_span']=year_span 
    df['genre'] =genre 
    df['run_time'] = run_time
    df['ratings'] =ratings 
    df['votes'] =votes
    return(df)



series('https://www.imdb.com/list/ls095964455/')
```

10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
def uci(url):
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    time.sleep(3)
    link=[x.get_attribute('href') for x in driver.find_elements_by_xpath("//td/span[@class='whitetext']/a")]
    driver.get(link[4])
    time.sleep(3)
    name = [x.text for x in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td")]
    df = pd.DataFrame({})
    df["Dataset_name"]=name[0::7]
    df['Data_type']=name[1::7]
    df['Task']=name[2::7]
    df['Attribute_type']=name[3::7]
    df['No_of_instances']=name[4::7]
    df['No of attribute']=name[5::7]
    df['Year']=name[6::7]
    df.drop(labels=0,axis =0, inplace =True)
    driver.close()
    return(df)

uci('https://archive.ics.uci.edu/')
```
