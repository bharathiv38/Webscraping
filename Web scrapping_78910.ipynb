```python
#importing libraries
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url=" https://www.flipkart.com/apple-iphone-11-black-64-gb-includesearpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace"
driver.get(url)
```

```python
search_field = driver.find_element_by_xpath("//input[@class='_3704LK']")
search_field
```

```python
Rating=[]
```

```python
for i in range(100):
    for j in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
        Rating.append(j.text)
    time.sleep(5)
```

```python
Rating
```

```python
for i in range(100):
    for j in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
        reviewsummary_tags.append(j.text)
    time.sleep(5)
```

```python
reviewsummary_tags
```

```python
Full_review=[]
```

```python
for i in range(100):
    for j in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
        Full_review.append(j.text)
    time.sleep(5)
```

Q10: Go to webpage https://www.amazon.in/
Enter “Laptop” in the search field and then click the search icon.
Then set CPU Type filter to “Intel Core i7” and “Intel Core i9” as shown in the below image:

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://www.amazon.in/'
driver.get(url)
```

```python
search_field=driver.find_elements_by_xpath("//input[@id='twotabsearchtextbox']")
search_field
```

```python
search_field[0].send_keys("Laptop")
```

```python
search_button=driver.find_elements_by_xpath("//div[@class='nav-search-submit nav-sprite']/span")
search_button[0].click()
```

```python
title_tags=driver.find_elements_by_xpath("//h2")
title_tags
```

```python
title_names=[]

for i in title_tags:
    title_names.append(i.text)
title_names
```

```python
price_tags=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
price_tags
```

```python
price_names=[]

for i in price_tags:
    price_names.append(i.text)
price_names
```

```python
Ratings_tags=driver.find_elements_by_xpath("//span[@class='a-size-base']")
Ratings_tags
```

```python
Ratings_names=[]

for i in Ratings_tags:
    Ratings_names.append(i.text)
Ratings_names
```

```python
import pandas as pd
table=pd.DataFrame({})

table['title_names']=title_names[:9]
table['price_names']=price_names[:9]
table['Ratings_names']=Ratings_names[:9]

table
```
